```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,ax = plt.subplots()
ax.scatter(df['Sales'], df['Profit'])
ax.set_title('Superstore Dataset')
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
```

This scatter plot shows the relationship between Sales and Profit. Each point represents an order, helping us visualize if higher sales generally lead to higher profits.

```python
fig, ax= plt.subplots()
ax.hist(df['Region'])
ax.set_title('Region wise sales review')
ax.set_xlabel('Region')
ax.set_ylabel('Frequency')
```

Here, a histogram is used to visualize how many orders come from each region. It's a quick way to understand regional distribution.

```python
fig, ax = plt.subplots()
ax.hist(df['Segment'])
ax.set_title('Customer- Segment wise Sales Review')
ax.set_xlabel('Segment')
ax.set_ylabel('Frequency')
```

This histogram breaks down the sales based on customer segments (e.g., Consumer, Corporate, Home Office), showing which segment has the most orders.

```python
fig ,ax= plt.subplots()
ax.hist(df['Category'])
ax.set_title('Category wise sale review')
ax.set_xlabel('Category')
ax.set_ylabel('Frequency')
```

```python
Similar to the previous plots, this one displays the number of orders by product category (e.g., Furniture, Office Supplies, Technology).
```

```python
fig, ax=plt.subplots()
ax.hist(df['Ship Mode'])
ax.set_title('Shipping mode wise sales review')
ax.set_xlabel('Ship Mode')
ax.set_ylabel('Frequency')
```

This plot shows how often different shipping modes are used, giving insights into customer preferences.

```python
df1= df[(df['Quantity']>=5)& (df['Profit']<30)]
sns.boxplot('Quantity', 'Profit', data=df1)
```

A box plot visualizes the distribution of profits for orders where the quantity is 5 or more, and the profit is less than $30. It helps identify outliers and the spread of profit values.

```python
sns.pairplot(df)
```

Finally, a pair plot is used to visualize the relationships between multiple numerical columns in the dataset. This is a great way to spot trends and correlations across different variables.
