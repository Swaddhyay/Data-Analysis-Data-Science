```python
df['ocean_proximity'].value_counts().plot(kind='bar', edgecolor='k', alpha = 0.7)
plt.xlabel("Area", fontsize=10)
plt.ylabel("Houses", fontsize=10)
plt.xticks(rotation=0)
plt.title("Houses Across the state of California (CA)", fontsize=15)
plt.show()
```

Purpose: 
This bar plot shows the distribution of houses based on their proximity to the ocean. The ocean_proximity column divides areas into "NEAR BAY," "INLAND," "NEAR OCEAN," "ISLAND," and "<1H OCEAN."

Insights: 
By visualizing the number of houses in each category, we can determine which areas are the most densely populated. For example, a higher "INLAND" bar indicates that more homes are located inland rather than near the ocean or bay. This information can help you understand the relationship between housing density and proximity to water bodies.


```python
df['median_house_value'].hist(bins=100)
plt.xlabel("Median House Value")
plt.ylabel("Houses")
plt.xticks(rotation=0)
plt.title("Median House value the statue of California (CA)")
plt.show()
```

Purpose:
The histogram depicts the distribution of median house values across different areas of California. The median_house_value column displays the median value of houses in various locations.

Insights: This chart can help you identify the range in which most house values fall. For example, if the histogram shows a peak in a specific range, it means that the majority of houses' values fall within that range. A right-skewed distribution may indicate that, while the majority of houses are moderately priced, there are some that are extremely valuable.

```python
sns.set(style="white")
sns_plot = sns.jointplot(x=df.latitude.values, y=df.longitude.values, height=10)
sns_plot.set_axis_labels("Latitude", "Longitude")
sns_plot.ax_joint.grid(True)
plt.show()
```

Purpose: 
This scatter plot shows the geographic distribution of houses based on their latitude and longitude. It's a visual representation of where houses are located across California.

Insights: 
The joint plot allows us to see concentrations of houses in specific regions. For example, a dense cluster of points in certain areas may indicate urban centers or coastal regions with high housing density. This visualization is helpful for understanding geographic trends and patterns in housing distribution.

```python
axes = df.hist(figsize=(20, 20))
for ax in axes.flatten():
    ax.title.set_fontsize(20)  
plt.show()
```

```python
sns_plot= sns.jointplot(x=df.population.values, y= df.median_house_value.values, height=10, alpha= 0.7)
plt.ylabel ("house_value")
plt.xlabel("population")
sns_plot.ax_joint.grid(True)
plt.show()
```

Purpose: This joint plot examines the relationship between population size and median house value. It plots population on the x-axis and median_house_value on the y-axis.

Insights: By analyzing this plot, we can identify if there is any correlation between population density and house values. For example, a positive correlation would suggest that areas with higher populations tend to have higher median house values, possibly due to increased demand in densely populated areas. Conversely, no clear pattern might indicate other factors influencing house prices.
